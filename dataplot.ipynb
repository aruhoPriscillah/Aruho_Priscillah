```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data.csv')  

df.head() 
```

```python
plt.figure(figsize=(7, 5))
sns.histplot(df['Duration'], kde=True, color='steelblue', bins=30)
plt.title('Duration Distribution')
plt.xlabel('Duration')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(7, 5))
sns.histplot(df['Pulse'], kde=True, color='steelblue', bins=30)
plt.title('Pulse Distribution')
plt.xlabel('Pulse')
plt.ylabel('Count')
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Duration Distribution
sns.histplot(df['Duration'], kde=True, ax=axes[0], color='steelblue', bins=30)
axes[0].set_title('Duration Distribution')
axes[0].set_xlabel('Duration')
axes[0].set_ylabel('Count')

# Pulse Distribution
sns.histplot(df['Pulse'], kde=True, ax=axes[1], color='steelblue', bins=30)
axes[1].set_title('Pulse Distribution')
axes[1].set_xlabel('Pulse')
axes[1].set_ylabel('Count')

# Overall title
plt.tight_layout()
plt.show()
```
